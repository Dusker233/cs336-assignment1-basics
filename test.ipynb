```python
from jaxtyping import Float
from einops import einsum, reduce, rearrange
import torch
```

```python
x: Float[torch.Tensor, "batch conv seq1 hidden"] = torch.ones(2, 2, 3, 4)
y: Float[torch.Tensor, "batch conv seq2 hidden"] = torch.ones(2, 2, 3, 4)

z = x @ y.transpose(-2, -1) # batch seq1 seq2
# 等价于使用 einops 中的 einsum
z1 = einsum(x, y, "batch conv seq1 hidden, batch conv seq2 hidden -> batch conv seq1 seq2")
# 也可以用 ... 省略不参与运算的维度
z2 = einsum(x, y, "... seq1 hidden, ... seq2 hidden -> ... seq1 seq2")

print(z, z1, z2)
```

```python
x: Float[torch.Tensor, "batch seq total_hidden"] = torch.ones(2, 3, 8)
w: Float[torch.Tensor, "hidden1 hidden2"] = torch.ones(4, 4)

x = rearrange(x, "... (heads hidden1) -> ... heads hidden1", heads=2)
x = einsum(x, w, "... hidden1, hidden1 hidden2 -> ... hidden2")

x = rearrange(x, "... heads hidden2 -> ... (heads hidden2)")
x.shape, x
```
